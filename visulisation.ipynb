### 4 Groups of Provinces

```python
def plot_group_135(group_provs, group_eng, prov_names, scores, int_dec):
    print(group_eng)

    int_dec = np.array(int_dec)
    prov_id_dict = {p:i for i, p in enumerate(prov_names)}
    prov_index = np.array([prov_id_dict[p] for p in group_provs])
    
    fig = plt.figure(figsize=(18,18)) # Create matplotlib figure

    scores = np.array(scores)
    ax = fig.add_subplot(111) # Create matplotlib axes
    
    b1 = ax.bar([2,6,10], scores[0][prov_index], label="Model 1")
    b3 = ax.bar([3,7,11], scores[1][prov_index], label="Model 2")
    b2 = ax.bar([1,5,9], int_dec[prov_index], label="True Value")
    
    plt.xticks([2,6,10], group_eng)
    plt.legend([b2, b1, b3], ["True Value", "Model 1", "Model 2"], loc="best", fancybox=True, framealpha=1, shadow=True, borderpad=1, fontsize="small")
    plt.grid(axis="y")
#     plt.ylim(0,0.25)
    plt.axhline(y=0)
    
    if not SAVE_FLAG:
        plt.show()
    else:
        plt.savefig(os.path.join("Figs", ",".join(group_eng) + "-135.png"), dpi=300, format='png')
```

```python
# plot_group_135(group_1, group_1_eng, prov_19, 
#                 [model_1_result["13th"].values, model_2_result["13th"].values], int_dec_rate)
plot_group_135(group_2, group_2_eng, prov_19, 
                [model_1_result["13th"].values, model_2_result["13th"].values], int_dec_rate)
plot_group_135(group_3, group_3_eng, prov_19, 
                [model_1_result["13th"].values, model_2_result["13th"].values], int_dec_rate)
plot_group_135(group_4, group_4_eng, prov_19, 
                [model_1_result["13th"].values, model_2_result["13th"].values], int_dec_rate)
```
